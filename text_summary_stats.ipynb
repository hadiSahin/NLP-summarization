```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume you have a DataFrame named df with a column named "resolution"
df = pd.read_csv("data/resolutions.csv")

# Count the number of words in each row of the "resolution" column
df['word_count'] = df['resolutions'].str.split().str.len()

# Plot the distribution of word counts
df['word_count'].plot.hist(bins=100)
plt.xlabel('Word Count')
# plt.show()
plt.savefig('data/graphs/word_count.pdf', format='pdf')
```

```python
median_word_count = df['word_count'].median()
print(median_word_count)
average_word_count = df['word_count'].mean()
print(average_word_count)
```

```python
import matplotlib.pyplot as plt

data = pd.read_csv("data/results_fineTuning/BART/bart_loss.csv")
print(data.head())
```
